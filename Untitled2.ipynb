```python
import numpy as np
import sympy
import matplotlib.pyplot as plt 
import cmath
```

```python
N=2048
tau=0.1*10**(-3)
t = np.linspace(-5*tau, 5*tau, N, endpoint=True) # задаем значение t
# зададим прямоугольник
def rect(arr_x,tau):
    return np.heaviside(arr_x/tau+0.5,0.5)-np.heaviside(arr_x/tau-0.5,0.5)
# зададим экспоненту
def exp(arr_x):
    return np.exp((-arr_x-tau)/tau)*(np.heaviside(arr_x+tau,0.5)-np.heaviside(arr_x,0.5))
# зададим сигнал
def sign(arr_x):
    return exp(arr_x)-exp(arr_x-2*tau)
S=sign(t) # массив сигнала
print(S)
```

```python
fig, ax = plt.subplots(figsize=(10, 5), dpi=600)
ax.plot(t/10**(-3),S)
ax.set_xlim([-0.1, 0.3])
ax.grid()
plt.xlabel("t,мс")
ax.set_ylabel(u'S(t), В')
plt.title('График импульсного сигнала')
```
